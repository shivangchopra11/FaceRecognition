```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
face_cascade = cv2.CascadeClassifier('./haarcascade_frontalface_alt.xml')
```

```python
data_path = './data/Surii_Chopra_Full/'
photos = []
```

```python
print(os.listdir(data_path))
```

```python
img1 = cv2.imread('./data/Surii_Chopra_Full/Surii_Chopra_0004.jpg',)
plt.imshow(img1)
faces = face_cascade.detectMultiScale(img1, 1.3, 5)
print(len(faces))
```

```python
pics = os.listdir(data_path)
print(len(pics))
ctr=0
for img in pics:
    print(ctr)
    pic = cv2.imread('./data/Surii_Chopra_Full/'+img)
    faces = face_cascade.detectMultiScale(pic, 1.3, 5)
    print(len(faces),"len")
    faces = sorted(faces, key=lambda x: x[2]*x[3], reverse=True)
    for face in faces[:1]:
        x,y,w,h = face
        offset = 7
        face_section = pic[y-offset:y+h+offset,x-offset:x+w+offset]
        face_section = cv2.resize(face_section,(100,100))
        print(ctr)
        cv2.imwrite("./data/Surii_Chopra/Surii_Chopra_" + "%04d.jpg" % ctr, face_section)
        ctr+=1
```
